```python
#plot the BERT precision and Cosine Similarity precision from the csv file
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('COToutput.csv')
#create a new df with bert and cosine similarity precision
df2 = df[['BERT_Precision', 'Cosine_Similarity']]
#plot BERT and Cosine Similarity precision in the same plot
plt.plot(df2['BERT_Precision'], label='BERT Precision')

# Plot Cosine_Similarity
plt.plot(df2['Cosine_Similarity'], label='Cosine Similarity')

# Add a legend
plt.legend()

# Show the plot
plt.show()

```
